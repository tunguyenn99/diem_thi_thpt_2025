```python
def tach_ngoai_ngu_theo_ma(df, col_ma='Mã môn ngoại ngữ', col_diem='Ngoại ngữ'):
    """
    Với mỗi dòng, dựa vào mã môn ngoại ngữ (N1-N7), tạo cột tương ứng như 'Tiếng Anh', 'Tiếng Nga'... chứa điểm ngoại ngữ.
    """
    ma_map = {
        'N1': 'Tiếng Anh',
        'N2': 'Tiếng Nga',
        'N3': 'Tiếng Pháp',
        'N4': 'Tiếng Trung',
        'N5': 'Tiếng Đức',
        'N6': 'Tiếng Nhật',
        'N7': 'Tiếng Hàn'
    }

    # Tạo các cột mới rỗng
    for mon in ma_map.values():
        df[mon] = None

    # Gán điểm ngoại ngữ vào đúng cột tương ứng
    for ma, ten_mon in ma_map.items():
        df.loc[df[col_ma] == ma, ten_mon] = df.loc[df[col_ma] == ma, col_diem]

    return df
```

# Bước 4: Lưu trữ các kết quả

```python
import os
```

```python
# Tạo thư mục nếu chưa tồn tại
def make_dirs(base_folder, group_name):
    path = os.path.join(base_folder, group_name)
    os.makedirs(path, exist_ok=True)
    return path
```

```python
# Gán nhãn nhóm khối (A, B, C, D) từ mã tổ hợp
def group_name(ma_khoi):
    if ma_khoi.startswith("A"):
        return "Ban A"
    elif ma_khoi.startswith("B"):
        return "Ban B"
    elif ma_khoi.startswith("C"):
        return "Ban C"
    elif ma_khoi.startswith("D"):
        return "Ban D"
    else:
        return "Khac"
```

```python
def plot_pretty_barplot(df, col_name, save_path, color="skyblue"):
    import matplotlib.ticker as mticker
    
    # Ép kiểu dữ liệu sang float, lỗi thì NaN
    data = pd.to_numeric(df[col_name], errors='coerce').dropna()

    # Chia bin theo khoảng 1 điểm (vd: [0–1), [1–2), ...)
    bins = np.arange(0, 30.1, 1)
    labels = [f"{int(left)}–{int(right)}" for left, right in zip(bins[:-1], bins[1:])]
    data_binned = pd.cut(data, bins=bins, labels=labels, right=False)

    # Đếm số lượng theo từng khoảng
    value_counts = data_binned.value_counts().sort_index()
    x = value_counts.index.astype(str).tolist()
    y = value_counts.values

    # Thống kê mô tả
    count = int(data.count())
    mean = data.mean()
    median = data.median()
    std = data.std()
    min_, max_ = data.min(), data.max()

    # Vẽ barplot với màu được chỉ định
    plt.figure(figsize=(18, 6))
    ax = sns.barplot(x=x, y=y, color=color, edgecolor="black")

    # Tiêu đề và trục
    plt.title(f"Phân bố tổng điểm: {col_name} ({count:,} thí sinh)", fontsize=14, weight='bold')
    plt.xlabel("Tổng điểm (theo khoảng)", fontsize=12)
    plt.ylabel("Số thí sinh", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.xticks(rotation=0, fontsize=9)

    # Annotate từng cột
    for i, val in enumerate(y):
        if val > 0:
            ax.annotate(f'{val:,}', xy=(i, val), ha='center', va='bottom', fontsize=8)

    # Box thống kê
    stats_text = '\n'.join([
        f"Số thí sinh: {count:,}",
        f"Trung bình (mean): {mean:.2f}",
        f"Trung vị (median): {median:.2f}",
        f"Độ lệch chuẩn (std): {std:.2f}",
        f"Min: {min_:.2f}   Max: {max_:.2f}",
        f"Phân vị 25% (Q1): {data.quantile(0.25):.2f}",
        f"Phân vị 75% (Q3): {data.quantile(0.75):.2f}",
        f"IQR (Q3 - Q1): {(data.quantile(0.75) - data.quantile(0.25)):.2f}",
        f"Số thí sinh >= 24 điểm: {(data >= 24).sum():,}",
        f"Số thí sinh >= 27 điểm: {(data >= 27).sum():,}",
    ])
    props = dict(boxstyle='round', facecolor='white', edgecolor='gray')
    plt.gca().text(0.02, 0.95, stats_text, transform=plt.gca().transAxes,
                   fontsize=10, verticalalignment='top', bbox=props)

    # Lưu hình
    plt.tight_layout()
    plt.savefig(save_path, dpi=150)
    plt.close()
```

```python
# Mapping màu sắc theo từng nhóm khối (Ban)
group_colors = {
    "Ban A": "#3498db",   # Xanh dương
    "Ban B": "#2ecc71",   # Xanh lá
    "Ban C": "#e67e22",   # Cam
    "Ban D": "#9b59b6",   # Tím
    # Có thể bổ sung thêm nếu có Ban E, H, K, v.v.
}

base_folder = "sources/internal"

for ma_khoi, mon_list in tohop_dict_full.items():
    if ma_khoi in df_2025_union.columns:
        # Xác định thư mục nhóm khối (Ban A, B, C, D)
        group = group_name(ma_khoi)
        folder_path = make_dirs(base_folder, group)

        # Đường dẫn lưu file ảnh
        save_path = os.path.join(folder_path, f"{ma_khoi}.png")

        # Lấy màu theo tên nhóm, fallback = "skyblue"
        color = group_colors.get(group, "skyblue")

        # Vẽ và lưu biểu đồ
        plot_pretty_barplot(df_2025_union, ma_khoi, save_path, color=color)
        print(f"✅ Đã vẽ xong {ma_khoi} ({group}) → {save_path}")
```
